```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
def f1(x):
    return x**2 + 2*x
def f2(x):
    return x**2 - 2*x
def f3(x):
    return x**2 - 2
x = sp.Symbol('x')
```

```python
x1 = np.linspace(-1.2, 0.2, 100) 
plt.plot(x1, f1(x1), 'g', label=f"${sp.latex(f1(x))}$")
x2 = np.linspace(-0.2, 1.2, 100)
plt.plot(x2, f2(x2), 'b', label=f"${sp.latex(f2(x))}$")
x3 = np.linspace(-1.2, 1.2, 200)
plt.plot(x3, f3(x3), 'r', label=f"${sp.latex(f3(x))}$")
x1_ = np.linspace(-1, 0, 100)
x2_ = np.linspace(0,  1, 100)
plt.fill_between(x1_, f1(x1_), f3(x1_), color="lightgray")
plt.fill_between(x2_, f2(x2_), f3(x2_), color="lightgray")
plt.grid()
plt.legend()
plt.gca().set_aspect(1)
```

```python
# Flächeninhalt der grauen Fläche:
```

```python
x = sp.Symbol('x')
A1 = sp.integrate(f1(x), (x, -1, 0))
display(A1)
A2 = sp.integrate(f2(x), (x,  0, 1))
display(A2)
A3 = sp.integrate(f3(x), (x, -1, 1))
display(A3)
print("Gesamtflächeninhalt:", A1 + A2 - A3)
```
