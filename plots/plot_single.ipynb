```python
import numpy as np
import matplotlib.pyplot as plt

def plotte_funktion(formel, x_min=-10, x_max=10):
    """
    Plottet eine einzelne Funktion basierend auf einer Formel.

    Args:
        formel (str): Die mathematische Formel als Text (z.B. "x**2 - 2*x").
        x_min (float): Die untere Grenze der x-Achse.
        x_max (float): Die obere Grenze der x-Achse.
    """
    # Schritt 1: Erzeuge viele x-Punkte im gewünschten Bereich.
    # 400 Punkte sind meist genug für eine glatte Kurve.
    x = np.linspace(x_min, x_max, 400)

    try:
        # Schritt 2: Berechne die y-Werte für jeden x-Punkt.
        # eval() führt den Formel-String als Python-Code aus.
        # Vorsicht: Nur mit vertrauenswürdigen Formeln verwenden!
        y = eval(formel)
    except Exception as e:
        print(f"Fehler bei der Auswertung der Formel '{formel}': {e}")
        return

    # Schritt 3: Erstelle den Plot.
    plt.figure(figsize=(8, 6))  # Grösse des Fensters
    plt.plot(x, y)              # Zeichne die Linie

    # Schritt 4: Füge Beschriftungen und ein Gitter hinzu.
    plt.title(f"Graph von f(x) = {formel}")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.grid(True)              # Zeige ein Gitter im Hintergrund

    # Füge horizontale und vertikale Achsenlinien hinzu
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)

    # Schritt 5: Zeige das fertige Diagramm an.
    plt.show()


# --- ANWENDUNGSBEISPIEL ---
# Hier können Sie die Funktion ganz einfach verwenden.
if __name__ == '__main__':

    # Definiere deine Formel als Text
    meine_formel = "-x**2 -x + 2"

    # Rufe die Funktion auf, um die Formel im Bereich von -4 bis 4 zu plotten
    plotte_funktion(meine_formel, x_min=-4, x_max=4)

    # Ein weiteres Beispiel
    # plotte_funktion("np.sin(x) * x", x_min=-20, x_max=20)
```
