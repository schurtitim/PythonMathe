```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Funktionen direkt definieren
def f(x):
  """f(x) = cos(x) - x/2"""
  return np.cos(x) - x / 2

def f_prime(x):
  """f'(x) = -sin(x) - 0.5"""
  return -np.sin(x) - 0.5

# 2. Vier Schritte des Newton-Verfahrens mit einer Schleife berechnen
num_steps = 4
# Eine Liste, um alle x-Werte zu speichern, beginnend mit x0
approximations = [0.0]

# Die Schleife f√ºhrt die 4 Iterationen durch
x_current = approximations[0]
for _ in range(num_steps):
  x_next = x_current - f(x_current) / f_prime(x_current)
  approximations.append(x_next)
  x_current = x_next

# Ergebnisse ausgeben
print(f"Berechnung f√ºr {num_steps} Schritte:")
for i, x_val in enumerate(approximations):
    print(f"x{i} = {x_val}")


# 3. Einfache Skizze erstellen
# Daten f√ºr die rote Kurve f(x)
x_kurve = np.linspace(0, 2.5, 400)
y_kurve = f(x_kurve)

# Rote Kurve zeichnen
plt.plot(x_kurve, y_kurve, 'r-', label='f(x)')

# x-Achse als Referenz zeichnen
plt.axhline(0, color='black', linewidth=0.8)

# Alle Newton-Schritte (Tangenten) mit einer Schleife zeichnen
for i in range(num_steps):
  x_start = approximations[i]
  x_end = approximations[i+1]
  plt.plot([x_start, x_end], [f(x_start), 0], 'b-', alpha=0.8) # alpha f√ºr leichte Transparenz

# Titel und Gitter f√ºr bessere Lesbarkeit
plt.title('Newton-Verfahren (4 Schritte)')
plt.xlabel('x')
plt.ylabel('y')
# Zoomen wir etwas auf den interessanten Bereich, damit man die Konvergenz besser sieht
plt.xlim(0, 2.5)
plt.ylim(-2, 1.5)
plt.grid(True)
plt.legend()

# Plot anzeigen
plt.show()
```
