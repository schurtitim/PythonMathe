```python
import sympy as sp
from sympy.plotting import plot

def berechne_tangente(funktion, variable, punkt_x0):
    """
    Berechnet die Tangentengleichung f端r eine Funktion an einem gegebenen Punkt
    und gibt die Details sowie die Gleichung als SymPy-Ausdruck zur端ck.
    """
    # Berechne den y-Wert am Ber端hrpunkt
    y0 = funktion.subs(variable, punkt_x0)
    
    # Bilde die erste Ableitung
    f_prime = sp.diff(funktion, variable)
    
    # Berechne die Steigung m an der Stelle x0
    steigung = f_prime.subs(variable, punkt_x0)
    
    # Stelle die Tangentengleichung auf: y = m*(x - x0) + y0
    tangenten_gleichung = steigung * (variable - punkt_x0) + y0
    
    # Gib die Ergebnisse formatiert aus
    print(f"Tangente an f(x) = {funktion} am Punkt x = {punkt_x0}:")
    print(f"  - Ber端hrpunkt P: ({punkt_x0}, {y0})")
    print(f"  - Steigung f'({punkt_x0}): {steigung}")
    print(f"  - Tangentengleichung: y = {sp.expand(tangenten_gleichung)}") # sp.expand() f端r eine saubere Form
    
    return tangenten_gleichung

# =========================================================================
# --- ANWENDUNG AUF IHRE AUFGABEN AUS DEM BILD ---
# =========================================================================
if __name__ == "__main__":
    
    # Definiere die Variable
    x = sp.Symbol('x')

    # Liste der Aufgaben aus dem Bild: (Funktion, x-Wert)
    aufgaben = [
        (sp.sin(x), 0),
        (sp.exp(x), 0),
        (sp.log(x), 1), # sp.log ist der nat端rliche Logarithmus ln(x)
        (x**2 - 2*x - 3, 2)
    ]
    
    print("="*50)
    print("      BERECHNUNG DER TANGENTENGLEICHUNGEN")
    print("="*50)

    tangenten = []
    # Schleife durch alle Aufgaben und l旦se sie
    for i, (funktion, punkt) in enumerate(aufgaben, 1):
        print(f"\n--- Aufgabe {i} ---")
        tangente = berechne_tangente(funktion, x, punkt)
        tangenten.append(tangente)

    # --- Optional: Grafische Darstellung aller Ergebnisse ---
    print("\n" + "="*50)
    print("      GRAFISCHE DARSTELLUNG")
    print("="*50)
    print("Die Graphen der Funktionen und ihrer Tangenten werden angezeigt...")

    # Erstelle die Plots f端r jede Funktion und ihre Tangente
    # Wir verwenden sp.plot, das mehrere Funktionen entgegennehmen kann
    p1 = plot(aufgaben[0][0], tangenten[0], (x, -3.14, 3.14), title='f(x) = sin(x)', show=False)
    p1[1].line_color = 'red' # Mache die Tangente rot
    
    p2 = plot(aufgaben[1][0], tangenten[1], (x, -2, 2), title='g(x) = exp(x)', show=False)
    p2[1].line_color = 'red'

    p3 = plot(aufgaben[2][0], tangenten[2], (x, 0.1, 4), title='h(x) = ln(x)', show=False)
    p3[1].line_color = 'red'

    p4 = plot(aufgaben[3][0], tangenten[3], (x, -1, 5), title='k(x) = x族-2x-3', show=False)
    p4[1].line_color = 'red'
    
    # Zeige die Plots in einem 2x2 Gitter an (ben旦tigt Matplotlib)
    from sympy.plotting import PlotGrid
    PlotGrid(2, 2, p1, p2, p3, p4)
```
