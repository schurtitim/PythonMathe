```python
import sympy as sp
from sympy.plotting import plot

def loese_aufgabe_2():
    print("\n" + "="*50)
    print("                 Lösung für Aufgabe 2")
    print("="*50)

    x = sp.Symbol('x')

    # 1. Definiere die drei Funktionen
    y_oben = -x**2 + 2*x + 1
    y_unten1 = -x**2 + x + 1
    y_unten2 = -x**2 + 3*x - 1

    print("Obere Begrenzung:", y_oben)
    print("Untere Begrenzungen:", y_unten1, "und", y_unten2)

    # 2. Finde die relevanten Schnittpunkte
    # Schnittpunkt, an dem die "untere" Begrenzung wechselt
    schnitt_unten = sp.solve(sp.Eq(y_unten1, y_unten2), x)
    x1 = schnitt_unten[0] # Ergibt x = 1

    # Schnittpunkte der oberen Kurve mit den beiden unteren
    schnitt_links = sp.solve(sp.Eq(y_oben, y_unten1), x)
    x0 = schnitt_links[0] # Ergibt x = 0

    schnitt_rechts = sp.solve(sp.Eq(y_oben, y_unten2), x)
    x2 = schnitt_rechts[0] # Ergibt x = 2
    
    print(f"\nDie relevanten Schnittpunkte liegen bei x = {x0}, x = {x1} und x = {x2}.")
    print("Die Fläche muss in zwei Integralen berechnet werden: von 0 bis 1 und von 1 bis 2.")

    # 3. Definiere und berechne die beiden Teil-Integrale
    # Intervall 1: [0, 1], die untere Grenze ist y_unten1
    integrand1 = y_oben - y_unten1
    flaeche1 = sp.integrate(integrand1, (x, x0, x1))
    print(f"\nFläche im Intervall [{x0}, {x1}]: Integral von ({integrand1}) dx = {flaeche1}")
    
    # Intervall 2: [1, 2], die untere Grenze ist y_unten2
    integrand2 = y_oben - y_unten2
    flaeche2 = sp.integrate(integrand2, (x, x1, x2))
    print(f"Fläche im Intervall [{x1}, {x2}]: Integral von ({integrand2}) dx = {flaeche2}")

    # 4. Addiere die Flächen
    gesamtflaeche = flaeche1 + flaeche2
    
    print("\n--- ERGEBNIS ---")
    print(f"Der gesamte Flächeninhalt beträgt: {flaeche1} + {flaeche2} = {gesamtflaeche}")
    
    # 5. Optional: Plot zur Visualisierung
    print("\nErstelle Plot zur Visualisierung...")
    p = plot(y_oben, y_unten1, y_unten2, (x, -1, 3), show=False, legend=True)
    p[0].line_color = 'red'
    p[0].label = 'y_oben'
    p[1].line_color = 'blue'
    p[1].label = 'y_unten1'
    p[2].line_color = 'green'
    p[2].label = 'y_unten2'
    p.show()


# --- Anwendung auf Aufgabe 2 ---
if __name__ == "__main__":
    # Den Code für Aufgabe 1 zuerst ausführen...
    # ... dann den für Aufgabe 2
    loese_aufgabe_2()
```
